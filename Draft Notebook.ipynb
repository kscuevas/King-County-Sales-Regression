```python
#removing home id, no. of floors, heat_source, lat & long coordinates
df.drop(['id', 'floors', 'heat_source',
         'lat','long'], axis=1, inplace=True)

df
```

```python
# Create zip codes from available home addresses
df['zip'] = [x.split(',')[2][-5:] for x in df['address']]
df
```

```python
# we only want records from home in the king county area. thus, we're removing all zipcodes from homes provided in
# the database that dont start with '98'
#sourced from: https://www.zillow.com/browse/homes/wa/king-county/ 
df = df[df.zip.str.startswith(('98'))]

df['city'] = [x.split(',')[1][0:] for x in df['address']]

df
```

```python
#create sale year column from datetime date info
df['year_sold'] = pd.DatetimeIndex(df['date']).year.astype(int)

df['year_sold'].value_counts()
```

```python
# creating column to figure out the age of the house
df['home_age'] = (df['year_sold']+1) - df['yr_built']

#creating a column to see when the home was last renovated (if applicable)
df['yr_from_reno'] = (df['year_sold']+1) - df['yr_renovated']

# creating a filter for yr_from_reno to ID houses that HAVENT been renovated
reno_filter = df['yr_from_reno'] > 2000

df
```

```python
# create new column for design_age to see time since build if it was not renovated, or time since renovation if renovated
# first set to renovation age
df['design_age'] = df['yr_from_reno']
#then update columns where there was no renovation done to age of the home
df['design_age'].loc[reno_filter] = df['home_age']
df
```
